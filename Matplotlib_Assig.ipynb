Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

### Scatter plot is the type of graph which is generated by Matplotlib library. Here data is visualised by intersaction of points whom are shown on x-axis and y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y, c = '#F70BD4', s = 50, alpha = 0.8, marker = 'x')
plt.xlabel("randomly generated x-values")
plt.ylabel("randomly generated y-values")
plt.title("Scatter_plot")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
# Subplot() function is used to draw multiple grapls in one single figure.
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.title("Line 1")
plt.subplot(4,1,1)
plt.plot(x1,y1)

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.title("Line 2")
plt.subplot(4,1,2)
plt.plot(x2,y2)

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.title("Line 3")
plt.subplot(4,1,3)
plt.plot(x3,y3)

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.title("Line 4")
plt.subplot(4,1,4)
plt.plot(x4,y4)
plt.show()
```

```python
# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot line 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

# Plot line 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. It is used to visualise the data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit, width = 0.8)
plt.xlabel("company")
plt.ylabel("profit")
plt.title("PROFIT OF COMPANIES")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(2,1,1)
plt.boxplot(box1)

plt.subplot(2,1,2)
plt.boxplot(box2)
```
